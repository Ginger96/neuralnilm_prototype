```python
def get_data_for_single_day(dataset, metergroup, target_appliance, start):
    SAMPLE_PERIOD = 6
    start = pd.Timestamp(start).date()
    end = start + timedelta(days=1)
    dataset.set_window(start, end)
    X = metergroup.power_series_all_data(sample_period=SAMPLE_PERIOD)
    y = metergroup[target_appliance].power_series_all_data(sample_period=SAMPLE_PERIOD)
    maximum = X.max()
    def get_diff(data):
        data -= data.min()
        data /= maximum
        data = data.diff().dropna()
        return data
    
    def index_as_minus_one_to_plus_one(data):
        data = get_diff(data)
        index = data.index.astype(np.int64)
        index -= np.min(index)
        index = index.astype(np.float32)
        index /= np.max(index)
        return np.vstack([index, data.values]).transpose()
    
    return index_as_minus_one_to_plus_one(X), get_diff(y)
```

```python
def quantize(data):
    N_BINS = 100
    MID = N_BINS // 2
    out = np.zeros(shape=(len(data), N_BINS))
    for i, d in enumerate(data):
        hist, _ = np.histogram(d, bins=N_BINS, range=(-1, 1))
        where = np.where(hist==1)[0][0]
        if where > MID:
            hist[MID:where] = 1
        elif where < MID:
            hist[where:MID] = 1
        out[i,:] = hist
    return (out * 2) - 1
```
